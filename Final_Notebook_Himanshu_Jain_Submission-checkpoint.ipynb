## 1. Introduction and Background

### Motivating the Problem
The global wine industry is a multi-billion dollar market where quality certification plays a pivotal role in pricing and consumer trust. This project investigates whether physicochemical properties—measurable laboratory statistics such as acidity, sugar levels, pH, and alcohol content—can effectively predict the sensory quality of wine. By analyzing these objective metrics, we aim to build a statistical model that can serve as a proxy for human evaluation, potentially automating parts of the certification process.

### Research Questions
Our analysis is guided by the following key questions:
1.  **Differentiation**: Do red and white wines differ significantly in their fundamental chemical composition (specifically alcohol content)?
2.  **Prediction**: Can we accurately predict the quality rating of a wine based solely on its chemical attributes?
3.  **Key Drivers**: Which specific chemical properties (e.g., volatile acidity, sulphates) have the strongest positive or negative impact on perceived quality?

### Dataset Description
We utilize the **Wine Quality Dataset** from the UCI Machine Learning Repository. This dataset consists of two distinct subsets:
-   **Red Wine**: 1,599 samples
-   **White Wine**: 4,898 samples

Each sample represents a specific wine tested in the Vinho Verde region of Portugal. The dataset includes 11 physicochemical input variables:
1.  **Fixed Acidity**: Most acids involved with wine or fixed or nonvolatile.
2.  **Volatile Acidity**: The amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste.
3.  **Citric Acid**: Found in small quantities, citric acid can add 'freshness' and flavor to wines.
4.  **Residual Sugar**: The amount of sugar remaining after fermentation stops.
5.  **Chlorides**: The amount of salt in the wine.
6.  **Free Sulfur Dioxide**: The free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion.
7.  **Total Sulfur Dioxide**: Amount of free and bound forms of S02.
8.  **Density**: The density of water is close to that of water depending on the percent alcohol and sugar content.
9.  **pH**: Describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic).
10. **Sulphates**: A wine additive which can contribute to sulfur dioxide gas (S02) levels, which acts as an antimicrobial and antioxidant.
11. **Alcohol**: The percent alcohol content of the wine.

The target variable is **Quality**, a score between 0 and 10 given by human experts.

# 2. Methods and Results

## Data Loading and Cleaning

To perform a comparative analysis, we first load the red and white wine datasets separately. Since the original files use semi-colons (`;`) as delimiters, we specify this during the read process. We then add a categorical identifier column (`type`) to each dataset before merging them into a single comprehensive dataframe, `wine_data`. This unified structure allows us to analyze the wines both collectively and by type.

```r
# Load datasets
red_wine <- read.csv("winequality-red.csv", sep = ";")
white_wine <- read.csv("winequality-white.csv", sep = ";")

# Add a type column to distinguish between the two datasets
red_wine$type <- "Red"
white_wine$type <- "White"

# Combine into a single dataframe
wine_data <- rbind(red_wine, white_wine)
wine_data$type <- as.factor(wine_data$type)

# Preview the data structure and summary statistics
head(wine_data)
summary(wine_data)
```

## Exploratory Data Analysis (EDA)

Exploratory Data Analysis is crucial for understanding the underlying structure of our data before applying formal statistical models. We will focus on three key areas: the distribution of our target variable (Quality), the relationship between alcohol and quality, and the correlation matrix of all chemical properties.

### Technique 1: Visualizing Distributions and Correlations

**1. Distribution of Wine Quality**
First, we examine how quality ratings are distributed across the dataset. This helps us identify if the classes are balanced or if there is a skew towards average wines.

```r
# Distribution of Quality
ggplot(wine_data, aes(x = quality, fill = type)) +
  geom_bar(position = "dodge") +
  labs(title = "Distribution of Wine Quality Ratings", 
       subtitle = "Comparison between Red and White Wines",
       x = "Quality Score (0-10)", 
       y = "Count") +
  theme_minimal()
```

**Observation**: The bar chart reveals that the majority of wines, both red and white, receive average scores of 5 or 6. Very few wines achieve excellent scores (8 or 9) or very poor scores (3 or 4). This normal-like distribution suggests that predicting extreme quality values might be challenging due to the lack of training data for those cases.

**2. Alcohol Content vs. Quality**
Next, we investigate the relationship between alcohol content and quality using boxplots. Alcohol is often cited as a primary driver of body and mouthfeel in wine.

```r
# Boxplot of Alcohol vs Quality
ggplot(wine_data, aes(x = as.factor(quality), y = alcohol, fill = type)) +
  geom_boxplot() +
  labs(title = "Alcohol Content by Quality Rating", 
       x = "Quality Score", 
       y = "Alcohol (% by volume)") +
  theme_minimal()
```

**Observation**: The boxplots show a clear trend: as the quality score increases, the median alcohol content also tends to increase. This suggests a positive correlation between alcohol level and perceived quality. Interestingly, this trend appears consistent for both red and white wines.

**3. Correlation Matrix**
To identify potential multicollinearity and see which variables are most strongly related to one another, we generate a correlation matrix.

```r
# Correlation Matrix (Numeric columns only)
if (!require("dplyr")) library(dplyr)
numeric_cols <- wine_data %>% select(where(is.numeric))
cor_matrix <- cor(numeric_cols)
corrplot(cor_matrix, method = "color", type = "upper", tl.cex = 0.7, 
         title = "Correlation Matrix of Physicochemical Properties", mar=c(0,0,1,0))
```

## Statistical Analysis

### Technique 2: Hypothesis Testing (Two-Sample T-Test)

We observed in our EDA that red and white wines might have different chemical profiles. To rigorously test this, we perform a two-sample t-test comparing the alcohol content of the two groups.

**Hypothesis:**
-   **Null Hypothesis ($H_0$)**: There is no difference in the mean alcohol content between Red and White wines ($\mu_{red} = \mu_{white}$).
-   **Alternative Hypothesis ($H_A$)**: There is a significant difference in the mean alcohol content ($\mu_{red} \neq \mu_{white}$).

```r
# T-test for Alcohol content between Red and White wines
t_test_result <- t.test(alcohol ~ type, data = wine_data)
print(t_test_result)
```

**Interpretation of Results:**
The t-test output provides us with a **p-value**. If this p-value is less than our significance level of 0.05, we reject the null hypothesis. 

Given the large sample size, we expect the p-value to be extremely small, indicating a statistically significant difference. The output also provides the **95% confidence interval** for the difference in means. If this interval does not include zero, it further confirms that the true population means are different. Practically, this tells us that the fermentation processes or grape varieties used for red and white wines result in distinct alcohol profiles.

### Technique 3: Bootstrapping for Difference in Means

To robustly estimate the difference in quality between Red and White wines without assuming a normal distribution, we will use **Bootstrapping**.

**Methodology:**
1.  Resample the quality ratings for Red and White wines separately with replacement.
2.  Calculate the mean quality for each resampled group.
3.  Compute the difference in means (White - Red).
4.  Repeat this process 1,000 times to build a distribution of the difference.
5.  Calculate the 95% Confidence Interval.

```r
# Bootstrapping Function
set.seed(123)
n_iterations <- 1000
diff_means <- numeric(n_iterations)

red_quality <- wine_data %>% filter(type == "Red") %>% pull(quality)
white_quality <- wine_data %>% filter(type == "White") %>% pull(quality)

for (i in 1:n_iterations) {
  sample_red <- sample(red_quality, length(red_quality), replace = TRUE)
  sample_white <- sample(white_quality, length(white_quality), replace = TRUE)
  diff_means[i] <- mean(sample_white) - mean(sample_red)
}

# Calculate Confidence Interval
ci_lower <- quantile(diff_means, 0.025)
ci_upper <- quantile(diff_means, 0.975)

print(paste("95% CI for Difference (White - Red): [", round(ci_lower, 4), ", ", round(ci_upper, 4), "]", sep=""))

# Plot Bootstrap Distribution
hist(diff_means, main="Bootstrap Distribution of Difference in Means", xlab="Difference (White - Red)", col="lightblue", border="white")
abline(v = c(ci_lower, ci_upper), col="red", lwd=2, lty=2)
```

**Interpretation:**
The histogram shows the distribution of the difference in mean quality between White and Red wines across 1,000 simulated samples. The red dashed lines represent the 95% Confidence Interval. If this interval does not contain zero, we can conclude with 95% confidence that there is a significant difference in quality between the two wine types.
